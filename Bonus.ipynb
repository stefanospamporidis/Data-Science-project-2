υλοποιήσω ήταν να ενώσω το τίτλο και το περιεχόμενο των άρθρων το όποιο ενδεχομένως να βελτίωνε ακόμα περισσότερο 
την απόδοση των αλγόριθμων.

```python
import csv
import numpy as np
import os, glob
import pandas as pd
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.linear_model.logistic import LogisticRegression
from sklearn.model_selection import train_test_split, cross_val_score
from sklearn import preprocessing
from string import punctuation
import matplotlib.pyplot as plt
import string
import nltk
from nltk.corpus import stopwords


my_stopwords = set(stopwords.words('english'))
other_words = {'said','say','says','just','did','was','were','year','years','like','people','tell','tells'
               'told','report','reported','get','also','however','already','may','almost','hes','ask','would',
              'could','one'}
my_stopwords = my_stopwords.union(other_words)
```

```python
#ka8arizei ena text sumvola, an einai adeio to kanei null
def clean_string(text): # ta kanei null
    text = ''.join([word for word in text if word not in string.punctuation])
    text = text.lower()
    text = ' '.join([word for word in text.split() if word not in my_stopwords])
    if text == '': 
        text = 'NULL' 
    return text
```

```python
data = pd.read_csv('data.tsv',sep="\t",encoding='utf-8')
updated_data = pd.DataFrame(data)
updated_data
```

```python
#remove punctuation and stopwords 
for i in range(0,len(updated_data['content'])):
    temp = clean_string(updated_data['content'][i])
    updated_data['content'].replace(to_replace=updated_data['content'][i],value=temp,inplace=True)

updated_data.to_csv('updated_data.tsv' ,sep='\t', quoting=csv.QUOTE_NONE ,index=False)
```
